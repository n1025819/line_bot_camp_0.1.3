```python
'''

讀取安全檔案內的字串，以供後續程式碼調用

'''
import json
secretFileContentJson=json.load(open("./line_secret_key",'r'))

print(secretFileContentJson.get("channel_access_token"))
print(secretFileContentJson.get("secret_key"))
print(secretFileContentJson.get("self_user_id"))
```

```python
'''

將先前準備的菜單設定檔，以Post消息寄發給Line

    設定Line的遠端位置
    設定消息的基本安全憑證
    寄發消息，並取得回應之Id
    

'''

import requests
menuRawData=[menuRawData1,menuRawData2]
menuJson1=json.loads(menuRawData[0])
menuJson2=json.loads(menuRawData[1])
createMenuEndpoint = 'https://api.line.me/v2/bot/richmenu'
createMenuRequestHeader={'Content-Type':'application/json','Authorization':'Bearer %s' % secretFileContentJson["channel_access_token"]}

#print(createMenuRequestHeader)

lineCreateMenuResponse = requests.post(createMenuEndpoint,headers=createMenuRequestHeader,data=json.dumps(menuJson1))
lineCreateMenuResponse1 = requests.post(createMenuEndpoint,headers=createMenuRequestHeader,data=json.dumps(menuJson2))
print(lineCreateMenuResponse)
print(lineCreateMenuResponse.text)
print(lineCreateMenuResponse1)
print(lineCreateMenuResponse1.text)
```

```python
'''

將先前準備的菜單照片，以Post消息寄發給Line
    取得上面設定檔的Id
    設定Line的遠端位置
    設定消息的基本安全憑證
    上傳照片，並取得回傳成果

'''

# 取得菜單Id 
uploadRichMenuId=json.loads(lineCreateMenuResponse.text).get("richMenuId")
uploadRichMenuId1=json.loads(lineCreateMenuResponse1.text).get("richMenuId")
print("貼到line_secret_key.rich_menu_id:", uploadRichMenuId)
print("貼到line_secret_key.rich_menu_id:", uploadRichMenuId1)
#'https://api.line.me/v2/bot/richmenu/{richMenuId}/content'

# 設定Line的遠端位置
uploadMenuEndpoint='https://api.line.me/v2/bot/richmenu/%s/content' % uploadRichMenuId
uploadMenuEndpoint1='https://api.line.me/v2/bot/richmenu/%s/content' % uploadRichMenuId1
print(uploadMenuEndpoint)
print(uploadMenuEndpoint1)
# 設定消息的基本安全憑證
uploadMenuRequestHeader={'Content-Type':'image/jpeg','Authorization':'Bearer %s' % secretFileContentJson["channel_access_token"]}

# 上傳照片
uploadImageFile1=open("./images/richmenu1.jpg",'rb')
lineUploadMenuResponse1=requests.post(uploadMenuEndpoint,headers=uploadMenuRequestHeader,data=uploadImageFile1)

uploadImageFile2=open("./images/richmenu2.jpg",'rb')
lineUploadMenuResponse2=requests.post(uploadMenuEndpoint1,headers=uploadMenuRequestHeader,data=uploadImageFile2)

with open('line_secret_key','r') as f:
    data = json.load(f)  
    data['rich_menu_id'] = uploadRichMenuId
    print(data)
with open('line_secret_key', "w") as f:
    f.write(json.dumps(data))

with open('line_secret_key','r') as f:
    data = json.load(f)  
    data['rich_menu_id_2'] = uploadRichMenuId1
    print(data)
with open('line_secret_key', "w") as f:
    f.write(json.dumps(data))

print(lineUploadMenuResponse1)
print(lineUploadMenuResponse1.text)


print(lineUploadMenuResponse2)
print(lineUploadMenuResponse2.text)
```
